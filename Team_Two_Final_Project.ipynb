```python
# Define target and feature variables
X = df[["What is your gender?", "What is your age?", "What country do you work in?", "Do you work remotely?", "Does your employer provide mental health benefits as part of healthcare coverage?", "If you have a mental health issue, do you feel that it interferes with your work when being treated effectively?", "If you have a mental health issue, do you feel that it interferes with your work when NOT being treated effectively?", "Have you heard of or observed negative consequences for co-workers who have been open about mental health issues in your workplace?", "If yes, what condition(s) have you been diagnosed with?"]].drop(columns="Does your employer provide mental health benefits as part of healthcare coverage?").values
y = df["Does your employer provide mental health benefits as part of healthcare coverage?"].values
```

```python
# Encode target values to binary

def encode_answers(answer):
     if answer == "No":
         return 0
     elif answer == "Yes":
         return 1
     elif answer == "I don't know":
         return 2
     elif answer == "Not eligible for coverage / N/A":
         return 3
     elif answer == "nan":
         return 4 
  

    
df["Does your employer provide mental health benefits as part of healthcare coverage?"] = df["Does your employer provide mental health benefits as part of healthcare coverage?"].apply(encode_answers)
df["Does your employer provide mental health benefits as part of healthcare coverage?"].unique()
```
